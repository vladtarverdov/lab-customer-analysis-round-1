```python
import pandas as pd 
import numpy as np
```

```python
file1 = pd.read_csv('file1.csv')
file1.head()
```

```python
file2 = pd.read_csv('file2.csv')
file2.head()
```

```python
file3 = pd.read_csv('file3.csv')
file3.head()
```

```python
# Rearrange the columns in the dataframe as needed
print(file1.columns)
print(file2.columns)
print(file3.columns)
```

```python
#Concatenate the three dataframes

column_names = file1.columns
data = pd.DataFrame(columns=column_names)
data = pd.concat([data,file1, file2, file3], axis=0)
data.head()
```

```python
print(file1.columns)
print(file2.columns)
print(file3.columns)
print(data.columns)
```

```python
#Delete the column education and the number of open complaints from the dataframe.

data.drop(columns='Education',axis=1,inplace=True)
```

```python
print(data.columns)
```

```python
data.drop(columns='Number of Open Complaints',axis=1,inplace=True)
```

```python
print(data.columns)
```

```python
# Correct the values in the column customer lifetime value. 
#They are given as a percent, so multiply them by 100 and change dtype to numerical type.

data['Customer Lifetime Value'] = data['Customer Lifetime Value']*100
```

```python

data['Customer Lifetime Value'] =  pd.to_numeric(data['Customer Lifetime Value'], errors='coerce')
data.head()
```
